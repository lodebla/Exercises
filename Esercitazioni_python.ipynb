```python
import numpy as np
import scipy.linalg as linalg
import matplotlib.pyplot as plt

def f(x):
  return (1+x)**2 - (9+2*x)**7

def get_A(n,x):
  A = np.zeros((x.size, n+1))
  for i in range(x.size):
    for j in range(n+1):
      A[i,j] = x[i]**j
  return A

def solve(N, n):
  x = np.linspace(-5 , 5, N)
  y = f(x)
  A = get_A(n,x)
  U, s, Vh = linalg.svd(A)
  alpha = np.zeros(s.shape)
  for i in range(n+1):
    ui=U[:,i]
    vi=Vh[i,:]
    alpha = alpha + ((ui.T @ y) / s[i]) * vi
  fig, (ax1, ax2) = plt.subplots(2, figsize=(20,10))
  ax1.plot(x,y)
  ax2.plot(x,alpha)

  return alpha


alpha = solve(15,4)
```
